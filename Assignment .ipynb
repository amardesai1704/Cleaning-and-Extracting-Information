```python
import pandas as pd 
import numpy as np
import re
```

```python
fname=open('docc.txt')
college_code_stream=[]
college_code_name=[]
university_name=[]
for line in fname:
    if re.search('^\d{4} -',line):
        college_code_name.append(line.strip())
    elif re.search('^\d{9}.* -',line):
        college_code_stream.append(line.strip())
    elif re.search('Home University :',line):
        university_name.append(re.findall('Home University : (.+)',line)[0])        
```

# Finding Cut off Ranks

```python
fname=open("docc.txt","r")
cutoff_rank=[]
lines=[line.strip() for line in fname.readlines()]
for i in range(0,len(lines)):
    if re.search('^\d{9}.* -',lines[i]):
        if re.search('^GOPEN[HS]',lines[i+3]):
            cutoff_rank.append(re.findall('^I (\d+)',lines[i+4])[0])
        else:
            cutoff_rank.append(np.nan)
```

```python
len(cutoff_rank)
```

# College is associated with the college code.

```python
dic_college_code_name={}
for w in set(college_code_name):
    words=w.split('-')
    words=[word.strip() for word in words]
    dic_college_code_name[words[0]]=words[1]
dic_college_code_name
```

```python
len(set(college_code_name))
```

# Stream is associated with the code.

```python
dic_code_stream={}
for w in college_code_stream:
    words=w.split('-')
    words=[word.strip() for word in words]
    dic_code_stream[words[0][4::]]=words[1]
dic_code_stream
```

```python
len(college_code_stream)
```

# Dataset is created using the information obtained.

```python
code=[]
for w in college_code_stream:
    words=w.split('-')
    words=[word.strip() for word in words]
    code.append(words[0])
```

```python
df_college=pd.DataFrame(columns=['Code','Name','Stream','University','Cutoff_Rank'])
df_college
```

```python
df_college['Code']=code
```

```python
for w in range(0,df_college.shape[0]):
    df_college.loc[w,'Name']=dic_college_code_name[df_college.loc[w,'Code'][0:4]]
    df_college.loc[w,'Stream']=dic_code_stream[df_college.loc[w,'Code'][4::]]
    df_college.loc[w,'University']=university_name[w]
    df_college.loc[w,'Cutoff_Rank']=cutoff_rank[w]
```
